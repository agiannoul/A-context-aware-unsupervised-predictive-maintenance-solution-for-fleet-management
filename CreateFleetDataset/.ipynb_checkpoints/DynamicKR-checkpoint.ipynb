```python
def normalize(dfs, with_mean=True, with_std=True):
        if with_mean:
            dfs = [df - df.mean() for df in dfs]

        if with_std:
            stds = [df.std() for df in dfs]
            for std in stds: std[std==0] = 1
            dfs = [df / std for df, std in zip(dfs, stds)]

        return dfs
```

#Dynamic K and R.

Calculate the objective function for each k in order to calculate the "best"/"dynamic" R by maximization of the objective function. -> return k tuples with (objective function value, Dynamic R)

objective function:( Dm/m + Dstd/std )/ #anomalies
                    where Dm is the difference bettween mean with all samples and mean without anomalies
                    and Dstd is the difference bettween standar-deviation with all samples and standar-deviation                         without anomalies.

Selecte the argmax of the objective function from the k tuples.

```python
#dynRK
def dynRk(dataframes_list,w,shiftdays,window,k_range,kmax):
    start = datetime.strptime("2000-01-01 00:00:00", "%Y-%m-%d %H:%M:%S") + dtime.timedelta(days=w*shiftdays)
    end = start + dtime.timedelta(days=window)
    windowdf=dataframes_list[0][start:end]
    for i in range(1,len(dataframes_list)):
        windowdf=windowdf.append(dataframes_list[i][start:end])
    X=windowdf.to_numpy()
    print("Historic valus: "+str(len(X)))
    #####
    kmax=min(kmax,len(X)-2)
    knn = NearestNeighbors(n_neighbors=kmax, algorithm='auto')
    knn.fit(X)
    dists,iindexes=knn.kneighbors(X, kmax+1, return_distance=True)
    argmax_all_k=[]
    k_range=range(5,kmax)
    for k in k_range:
        kdists=[]
        for distsample in dists:
            kdists.append(distsample[k+1])
        meso = statistics.mean(kdists)
        diviation = statistics.stdev(kdists)
        z=[v/5 for v in range(1,50)] #z vector for threshold calculation
        r= [meso+(element * diviation) for element in z] # r: set of candidate R
        maxk=-1
        maxR=r[0]
        for R in r:
            #EA =[] #List of sequence of anomalous errors
            ea=[]# list of anomaly errors\
            for dd in kdists:
                if dd > R:
                    ea.append(dd)
            # if ea equals to 0 that means no anomalies so the Δμ/μ and Δσ/σ also are equal to zero
            if len(ea)==0:
                continue
            if len([element for element in kdists if element<=R])<=1:
                continue
            # Δμ -> difference betwen mean of errors and mean of errors without anomalies
            dmes=meso - statistics.mean([element for element in kdists if element<=R])
            # Δσ ->  difference betwen diviation of errors and diviation of errors without anomalies
            ddiv=diviation - statistics.stdev([element for element in kdists if element<=R])
            argmaxError=(dmes/meso+ddiv/diviation)/(len(ea))
            if argmaxError>maxk:
                maxk=argmaxError
                maxr=R
        #print(maxk,end=' ,')
        argmax_all_k.append((maxk,maxR,k))
    triplek = max(argmax_all_k, key=lambda e: e[0])
    finalk=triplek[2]
    finalR=triplek[1]
    return finalk,finalR
```
