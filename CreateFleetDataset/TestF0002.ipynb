# Proccess

Sort dataset using the end of life.
Split Dataset:
    Using predifine step value we split the dataset into |Data|/step:
    We selecte the first (with smallest life) timeseries and put it in the first bin, respectively Second timeseries to second bin and so on. Every bin will contain timeseries with at least step distance in sorted dataset. This is done in order to, end of life of each equipment in each bin is as far away possible.

Therefore in f001 fo example which cardinality is 100 and using step=5 we have four dataset (25 instances in each).

Set the rules:
    In order to built an realistic dataset for fleet Pdm and anomaly detection, we have to set some rules. In real life fleet cases, there a sufficient ammount of data considered normal and a small porpotion of data deviating each time. So, to fullfil this requirment we set two parameters in order to built our fleet dataset.


mindist: is the minimum distance in time (here cycles) between end of life of two instances a,b in the dataset.

Good Refernce: an instance b is consider good reference for a instance if
1)b starts before the end of life of a 
2)if b end of life is at least mindist time (cycles) after a end of life ( b.end-a.end>mindist )

Bad Reference: an instance b is consider bad reference for a instance if
1)b starts before the end of life of a
2)if b end of life is has not at least mindist time (cycles) difference with a end of life ( abs(b.end-a.end)>mindist ). Two instances are consdidered that deviates near in time


Minimum_Good_Porpotion_refference: is the minimum porpotion that an instance a must has in order to be consider valid for analysis. Porpotion of each instance is calculated between bad and good refernces ( #good/(#good+#bad) ).

Having as a base the first constacted Bbin we use the instances of the others bins (Rbins) in order to improve refference of first one. Specifically in repeated procedure:

1) find first instance of Bbin which is not foolfield the Good_Porpotion_refference -> targetInstance.
2) imporve targetInstance Good_Porpotion_refference:\
    Until targetInstance has Good_Porpotion_refference>Minimum_Good_Porpotion_refference:
    Get first instance (temp) from Rbins and shift it in time with shift=2*targetInstance.end/3
    // this is done in order to targetInstance has sufficient overlap with temp instance and 
    // end of temp instance is as far away as possible with targetInstance.
    add temp in Bbin.

```python
Buildtuples3=Buildtuples2
flat_list1 = [item for sublist in Buildtuples3[0] for item in sublist]# for item2 in item]
#print(Buildtuplesindexes2)
flat_listindexes1 = [item for sublist in Buildtuplesindexes2[0] for item in sublist]# for item2 in item]
print(flat_listindexes1)
flat_list1.sort()
#print(flat_list1)
pairlist1=[(0,i) for i in flat_list1]
#print(pairlist1)
```

```python
def refporpotion(flat_list1):
    normalrefs=[]
    #print(endes)
    mindist=60
    for item in flat_list1:
        counter=0
        for end in flat_list1:
            if end[0]<=item[1] and end[1]-item[1]>mindist:
                counter+=1
        normalrefs.append(counter)

    badrefs=[]
    for item in flat_list1:
        counter=0
        for end in flat_list1:
            if end[0]<=item[1] and   abs(end[1]-item[1])<=mindist:
                counter+=1
        badrefs.append(counter)

    dif_bad_normal=[]
    for i in range(len(badrefs)):
        if normalrefs[i]==0:
            good_bad_porpotion=0
        else:
            good_bad_porpotion=normalrefs[i]/(normalrefs[i]+badrefs[i])
        dif_bad_normal.append(good_bad_porpotion)
    return dif_bad_normal
```

```python
goodlimit=0.65
while len(pairlist2)>0:
    targetpos=-1
    c=0
    porpotion=refporpotion(pairlist1)
    for item in porpotion:
        if item<goodlimit:
            targetpos=c
            break
        c+=1
    if targetpos==-1:
        break
    targettuple=pairlist1[targetpos]
    goodrefs=normalrefs[targetpos]
    badref=badrefs[targetpos]
    toextend=[]
    used=[]
    isDone=False
    i=0
    while len(pairlist2)>i and isDone==False:
        temp=pairlist2[i]
        shift=targettuple[0]+int(2*targettuple[1]/3)
        temp=(temp[0]+shift,temp[1]+shift)
        if temp[0]<targettuple[1] and temp[1]-targettuple[1]>mindist:
            goodrefs+=1
            toextend.append(temp)
            used.append(i)
        if goodrefs/(goodrefs+badref)>goodlimit:
            isDone=True
        i+=1
    newpairlist1=[]
    newpairlist1.extend(pairlist1)
    newpairlist1.extend(toextend)
    print(used)
    newpairlist2=[]
    for i in range(len(pairlist2)):
        if i not in used:
            newpairlist2.append(pairlist2[i])
    pairlist1=newpairlist1
    pairlist2=newpairlist2
```
